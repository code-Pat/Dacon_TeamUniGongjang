```python
gyeongi_data = pd.read_csv('./CSV/population/for_graph2.csv', header = 0)
```

```python
req = requests.get('https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2013/json/skorea_municipalities_geo_simple.json')
state_geo = req.json()
```

```python
#print(state_geo)
```

```python
seoul_map = folium.Map(location = [37.5502, 126.982],
                      zoom_start = 10.5)
                      #tiles = 'cartodbpositron')
                      #tiles = 'openstreetmap')
folium.Choropleth(
    geo_data = state_geo,
    name = '서울시 시각화',
    data = seoul_data,
    columns = ['district_name','Y2021-Y2016'],
    key_on = 'feature.properties.name',
    fill_color = 'Blues',
    fill_opacity = 0.7,
    line_opacity = 0.3,
    color = 'gray',
    legend_name = 'growth rate'
    ).add_to(seoul_map)

folium.Choropleth(
    geo_data = state_geo,
    name = '경기도 시각화',
    data = gyeongi_data,
    columns = ['district_name', 'Y2021-Y2016'],
    key_on = 'feature.properties.name',
    fill_color = 'Greens',
    fill_opacity = 0.7,
    line_opacity = 0.3,
    color = 'gray',
    legend_name = 'growth rate'
    ).add_to(seoul_map)

folium.LayerControl(collapsed = True).add_to(seoul_map)


tooltip = "Click"

# B points Markers
folium.Marker(
    [37.6899062,127.0428593], 
    popup = "<b>B-1 도봉로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.6941317,127.0509975], 
    popup = "<b>B-2 동일로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.5544386,127.1083746], 
    popup = "<b>B-3 아차산로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.5947282,127.1102168], 
    popup = "<b>B-4 망우로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.6186095,127.1023901], 
    popup = "<b>B-5 경춘북로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.6186082,127.0871067], 
    popup = "<b>B-6 화랑로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.6111681,127.1124202], 
    popup = "<b>B-7 북부간선</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.5126834,127.1500208], 
    popup = "<b>B-8 서하남로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.5491794,127.177156], 
    popup = "<b>B-9 천호대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.5730652,127.1634344], 
    popup = "<b>B-10 올림픽대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.4634929,127.0375331], 
    popup = "<b>B-11 경부고속</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.4701094,127.1217181], 
    popup = "<b>B-12 송파대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.4647718,127.106082], 
    popup = "<b>B-13 밤고개로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.3718672,127.0997269], 
    popup = "<b>B-14 분당수서</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.464237,126.9869137], 
    popup = "<b>B-15 과천대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.4634929,127.0375331], 
    popup = "<b>B-16 양재대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.4698955,127.0111568], 
    popup = "<b>B-17 반포대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)

folium.Marker(
    [37.435161,126.9000497], 
    popup = "<b>B-18 시흥대로</b>", 
    tooltip = tooltip
).add_to(seoul_map)


```
